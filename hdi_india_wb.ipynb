```python
ind_hdi = df3.join(df4)
```

```python
ind_hdi.rename(columns={'Region':'State_UT'},inplace=True)
```

```python
state = pd.read_html('https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_area')
```

```python
region = state[1][['State (S) / Union territory (UT)','Region']]
```

```python
region.rename(columns={'State (S) / Union territory (UT)':'State_UT'},inplace=True)
```

```python
ind_hdi_melt.reset_index(inplace=True)
```

```python
ind_region_hdi = ind_hdi_melt.merge(region,how='left',on='State_UT')
```

```python
ind_region_hdi
```

```python
ind_region_hdi.isna().sum(axis=0)
```

```python
ind_region_hdi[ind_region_hdi.isna().any(axis=1)]['State_UT'].unique()
```

```python
ind_hdi_melt['State_UT'].unique()
```

```python
ind_region_hdi['Region'].value_counts(dropna=False)
```

```python
ind_region_hdi.set_index('State_UT',inplace=True)
```

```python
ind_region_hdi.loc['Chandigarth','Region'] = ind_region_hdi.loc['Chandigarth','Region'].fillna('Northern')
```

```python
ind_region_hdi.loc['Dadra and Nagar Haveli','Region'] = ind_region_hdi.loc['Dadra and Nagar Haveli','Region'].fillna('Western')
```

```python
ind_region_hdi.loc['Daman and Diu','Region'] = ind_region_hdi.loc['Daman and Diu','Region'].fillna('Western')
```

```python
ind_region_hdi.loc['New Delhi','Region'] = ind_region_hdi.loc['New Delhi','Region'].fillna('Northern')
```

```python
ind_region_hdi.loc['Orissa','Region'] = ind_region_hdi.loc['Orissa','Region'].fillna('Eastern')
```

```python
ind_region_hdi.loc['Uttaranchal','Region'] = ind_region_hdi.loc['Uttaranchal','Region'].fillna('Northern')
```

```python
ind_region_hdi.reset_index(inplace=True)
```

```python
ind_region_hdi['Region'].value_counts(dropna=False)
```

```python
ind_region_hdi.loc['Lakshadweep','Region'] = ind_region_hdi.loc['Lakshadweep','Region'].replace(to_replace='Western',value='Southern')
```

```python
ind_region_hdi.loc['Andaman and Nicobar Islands','Region'] = ind_region_hdi.loc['Andaman and Nicobar Islands','Region'].replace(to_replace='Bay of Bengal',value='Southern')
```

```python
ind_region_hdi.reset_index(inplace=True)
```

```python
ind_region_hdi.info()
```

```python
groupby_Region = ind_region_hdi.groupby(['Region','Year'])['Hdi'].mean()
```

```python
pd.options.display.max_rows = 200
```

```python
groupby_region = groupby_Region.to_frame()
```

```python
groupby_region.reset_index(inplace=True)
```

```python
groupby_region.set_index('Region',inplace=True)
```

```python
fig,ax = plt.subplots(figsize=(18,12))
plt.scatter(x=groupby_region.loc['Central','Year'],y=groupby_region.loc['Central','Hdi'],label='Central')
plt.scatter(x=groupby_region.loc['Northern','Year'],y=groupby_region.loc['Northern','Hdi'],label='Northern')
plt.scatter(x=groupby_region.loc['Northeastern','Year'],y=groupby_region.loc['Northeastern','Hdi'],label='Northeastern',marker='s')
plt.scatter(x=groupby_region.loc['Western','Year'],y=groupby_region.loc['Western','Hdi'],label='Western',color='Yellow')
plt.scatter(x=groupby_region.loc['Eastern','Year'],y=groupby_region.loc['Eastern','Hdi'],label='Eastern',color='red')
plt.scatter(x=groupby_region.loc['Southern','Year'],y=groupby_region.loc['Southern','Hdi'],label='Southern',marker='*',color='black')
ax.legend(fontsize='x-large')
ax.set_title('Human Devolopment Index, Indian State,Regionwise',fontsize=20,color='red')
ax.set_xlabel('Year',fontsize=15)
ax.set_ylabel('HDI Index',fontsize=15)
```
