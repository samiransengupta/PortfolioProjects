```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
bihar_ae_20 = pd.read_csv('bihar_ae_2020.csv') # data source - TCPD Ashoka Univercity
```

```python
pd.options.display.max_rows = 200
```

```python
bihar_ae_20.head(20)
```

```python
bihar_ae_20.info()
```

```python
bihar_ae_20 = bihar_ae_20[bihar_ae_20['Poll_No']==0]
```

```python
bihar_ae_20.columns
```

```python
find_spoiler = bihar_ae_20[['Constituency_No','Constituency_Name','Party','Position','Votes','Valid_Votes','Vote_Share_Percentage','Margin']]
```

```python
find_spoiler.tail(20)
```

```python
margin1 = find_spoiler.loc[find_spoiler['Position']==1,['Margin','Constituency_No']]
```

```python
votes3 = find_spoiler.loc[find_spoiler['Position']==3,['Votes','Constituency_No']]
```

```python
margin1.reset_index(drop='index',inplace=True)
```

```python
votes3.reset_index(drop='index',inplace=True)
```

```python
margin1['difference-Margin_of_Winner,position_3'] = margin1['Margin'] - votes3['Votes']
```

```python
spoilers = margin1[margin1['difference-Margin_of_Winner,position_3']<0]
```

```python
spoilers
```

```python
spoilers_constituency_list = spoilers['Constituency_No'].to_list()
```

```python
spoilers_df = find_spoiler.loc[find_spoiler['Constituency_No'].isin(spoilers_constituency_list)]
```

```python
spoilers_df.loc[spoilers_df['Position']==3,['Party','Constituency_Name']]['Party'].value_counts()
```

```python
ljp_spoilers_constituency_list = spoilers_df.loc[(spoilers_df['Party']=='LJP') & (spoilers_df['Position']==3)]['Constituency_No'].to_list()
```

```python
ljp_spoilers_constituency_list
```

```python
spoilers_df.loc[(spoilers_df['Constituency_No'].isin(ljp_spoilers_constituency_list))&(spoilers_df['Position']==2)]['Party'].value_counts()
```

```python
spoilers_df.loc[(spoilers_df['Constituency_No'].isin(ljp_spoilers_constituency_list))&(spoilers_df['Position']==2)]['Party'].value_counts().plot(kind='barh',figsize=(12,8))
plt.xlabel('Number of Seat',fontsize=18)
plt.title('Number of AC lost by Any Party due to LJP,Bihar Election 2020',fontsize=20,color='red')
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.savefig('ljp_spoilers_bihar_20.jpg')
plt.show()
```
